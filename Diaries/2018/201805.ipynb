```
年入百万的人，
都是怎么过日子的？记得前阵子，有碗汤火了：有钱人的快乐，你无法想象。

很多人或许以为，自己过得不开心，是钱不够。小姐姐讲几个年入百万朋友的故事，看是不是这样。

一 虽样本不够全，但有一定代表性，给他们按字母排个序。朋友A，90年，男孩子，工作五年，产品经理。

第一份工作进了美团（现在叫新美大），在大公司搬砖，其实是个有心人，工作只是一方面，另一个重点是混圈子，资本圈和币圈。

最近要跳槽，今日头条给了offer，年薪50万+50万的期权，这么一算，劳动收入过百万门槛，也算指日可待。

去年这家伙收入过百万了，不知他是什么时候买的比特币，去年的行情大家也都知道，算是牛市，他个人账户算上浮盈，大小加起来过了千万。

到了现在，陆续落袋为安，算起来也有几百万流动资金。不过生活照旧，过的普普通通，北京还没买房资格，现在还是租房住,其貌不扬，人群里也很普通，吃穿更不讲究，便利店十几块的盒饭，优衣库常年打折促销的T恤，都是必备款。

朋友B，88后，男，流量圈创业，央企上班，月入过百万，年入过千万。算是一个奇葩，他的项目去年被收购，套现了几千万，还继续在原公司上班。

而且这个项目，还是自己兼职投入、远程参与。

南方人头脑活、有毅力、赚钱动力，在他身上展现。

14年和几个朋友，硬是凑出100万，做了第一个小成本美甲店。

更厉害的，他是流量运营的高手，擅长效果转化，一个店的套路弄清，资金回本后，再去复制新店，两年多下来，团队在全国上百城市开店，现金流还不错
。

生活也很简单
，
衣食住行可以说很糙
，
甚至比不上年入十万的白领精致
。

那些所谓的精致生活
，
貌似离他都很远
。

没有买车
，
最近准备上海买房
，
全身上下是一股淘宝风
，
阿玛尼
、
爱马仕这种是什么
，
不存在的
。

C
，
85后
，
男
，
创业板医药上市公司CEO
，
年薪200万加股票
。

是朋友的朋友
，
不算股票的话
，
还可以淡定搬砖
。
算上股票
，
身家过了九位数
，
反而是战战兢兢
。

去年公司上市成功
，
身家暴涨
，
他说原来的那些亲戚
，
感觉突然变了
，
害怕冲着他的钱来的
。

坐标某二线
，
有别墅和住宅
，
一辆代步车
，
年入200万的时候
，
虽常加班
，
但生活还算美满
，
有位相亲认识
、
互相深爱的老婆
，
自己的职责是每天赚钱
，
老婆负责貌美如花加周末清迈游
。

突然有钱了
，
烦恼反而更多了
，
身边的亲戚朋友
，
反倒要花心思来平衡
。

朋友D
，
85后
，
男
，
BAT某家的中层
，
配期权
，
搬砖7年
，
工资加期权
，
税前接近百万
。

大公司搬砖多年
，
搬砖多年
，
公司早些年股价不高的时候
，
绩效考核给发期权
。

他心比较大
，
美股账户一直懒得打理
，
也就没卖出
，
现在单算这些期权
，
行权以后
，
估计有一百多万能进自己腰包
。

不过
，
他虽算打工里收入较高的
，
生活压力可不小
。

人在上海
，
买房不算早
，
每个月房贷2万多
，
如果没有老婆的收入进来
，
估计生活紧巴巴
。

近几年身体健康大不如前
，
有了中年人的恐慌
，
去年刚给家人配置保险
。

还要考虑孩子留学
、
父母养老
，
算是这里面压力最大的一位了
。

E
，
东部二线
，
男
，
70后
，
大学教授
，
接外快项目
，
年入百万
，
但收入逐年下滑
。

这位是朋友的大学老师
，
朋友是资源勘探类的
，
这位老师自己有个公司
，
工作多年
，
科研没怎么搞
，
项目倒不少做
。

背靠院校资源
，
桃李遍布行业大小公司
，
几十年深耕下来
，
才有了这样的局面
。

不过
，
积累了多年资源
，
也面临新能源的冲击
，
原来的老路子越走越窄
，
这些年明显收益下滑
。

这位老师
，
家境一般
，
经历过穷日子
，
虽然收入上来了
，
可得很节俭
，
最大的开销是抽烟
，
但也都是普通的牌子
。

有一辆十几万的普通代步车
，
用来出差跑项目
。

不过
，
这几年房子倒没少买
，
不懂投资
，
但跟上了城市楼市的涨幅
。

二

他们的日子很普通
，
无非现金多一点
，
花钱时候更从容
。

但吃穿用度
，
和大家没多大差别
。

对于他们
，
一顿400多的日料
，
有的人会舍不得
，
更别提过得像前阵子的隐形贫困人口：

能买戴森吸尘器就不用扫帚了；吃完牛油果又要吃藜麦了；100块钱一张的“前男友面膜”用起来也不心疼；一有健身冲动
，
就非得去办张年卡
。

要过得没焦虑
，
有很强的安全感
，
还是达不到的
。

对于他们
，
在一线城市有房没房
，
差距还是挺大的
。
买房晚的
，
哪怕现在收入看着高
，
生活压力并不小
。

但也有很多值得敲黑板
，
我发现他们有些共同点：

收入的增长
，
不是一蹴而就
，
而是背后的积累
，
有的靠人脉变现
，
有的是行业资源
，
有的事物洞察
，
有的是投资风口
。

当然
，
也有单纯搬砖像朋友D这种
，
往往金融圈
、
互联网圈
、
科技圈居多
。

传统行业
，
很难达到
，
收入的上限也摆在那里
。

他最近也是在看项目
，
找机会准备创业
，
或者加入一个创业团队
。

另外
，
他们往往好奇心强
，
执行力也到位
。

看到一个事
，
能够独立思考
，
或者再去请教别人
，
找到背后的原因
。

学习能力普遍更高
，
往往认准了
，
下次再和你聊
，
已经开始做
，
或遇到了某个问题
。

三

年入百万只是努力的一个表面
，
但不是生活本身
。

真正美好的东西
，
往往花钱买不到
，
也不能用钱衡量
。

比如每天早起的阳光
，
一段冥想
，
靠的是自律
，
和钱无关
。

比如温馨的家庭
，
懂事的孩子
，
体贴的另一半
，
亲人的孜孜温情...

咱们这人多
，
难免有人抬杠
，
说你自己有没有年入百万
，
不要在这啰嗦
。

这也是我另外想说的
，
用赚多少钱去肯定一个人
，
其实很幼稚
，
或者说肤浅
。

健全的人品
，
内心的善良
，
自律的生活态度
，
洋溢的热情
，
才是宝贵的
，
也不是钱能衡量的
。

如果单纯看钱
，
咱们这个号功利一点
，
多去接点广告
，
垃圾平台
，
垃圾硬广
，
一个月就能达到了
。

可我不想做
，
很多时候
，
钱只是一方面
，
真正有价值的东西
，
往往是钱买不到的
，
比如彼此的信任
。

最近一个感悟
，
做事不应该用赚多少钱做目标
，
否则容易心浮气躁
，
而要关注事情本身
，
给对方带来价值
。

下一个阶段
，
我在想该怎么给大家带来价值
，
想了几天脑核疼
，
居然比聊干货还费力
。

好了
，
交给你了
，
留言唠唠
，
说说你的想法


作者：理财的我
链接：https://xueqiu.com/1169501454/106661143
来源：雪球
著作权归作者所有。商业转载请联系作者获得授权，非商业转载请注明出处。
```
