#### Python 多线程，

GIL: 理解Python里面的多线程，为什么是鸡肋，为什么有时候可以提高效率，为什么有时候又不行？一个Python进程由于又Global Interpretor Lock的存在，线程只能在获取这个排他锁的时候才能运行，导致在现代的多核心多线程的机器上，一个Python进程的线程是分时进行的，而不是并发进行，所以对于计算密集型的程序，多线程并没有多少的提升，本质还是只用到了一个处理器，没有最大化利用硬件;但是对于IO密集型的进程，大部分线程在等待IO的时候，这时候不会阻塞进程，因为可以运行的线程会一直运行，因此提高了效率，更多更准确要Google, StackOverflow去把。。。

- 单线程： 读数据 - 处理数据 - 读数据 循环; 730seconds;

- 双线程： 读数据 - 处理数据/同时开始读数据; 632seconds;

启动两个线程之后，时间应该接近取数时间了，Python用于开发原型，测试，的确很合适，原型满意了就可以使用效率更高的编译型语言了。

表空间： 
是一个或多个数据文件的集合，所有的数据对象都存放在指定的表空间中，但主要存放的是表， 所以称作表空间。 

分区表： 
当表中的数据量不断增大，查询数据的速度就会变慢，应用程序的性能就会下降，这时就应该考虑对表进行分区。表进行分区后，逻辑上表仍然是一张完整的表，只是将表中的数据在物理上存放到多个表空间(物理文件上)，这样查询数据时，不至于每次都扫描整张表。
