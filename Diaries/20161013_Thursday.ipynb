#### Important Message Form XFL
作业分为了 DataStage作业 和 PROC存储过程; moia是通过Datastage的api来调动DS作业的;通过shell连接数据库调动proc作业。因此配置的作业名不同，ds是作业名，proc是shell名; 另外要注意作业之间的依赖关系。

一般来说，莫个作业要用到那些表，就要等待那些表对应的作业先完成（**数据结构的拓扑结构，工程里面的先后顺序**），你的作业就要依赖那些作业，像SDM的作业就要依赖STA的作业完成，FDM的作业要依赖SDM的作业完成，FDM的历史表就要依赖FDM的源作业完成

STA层的moia配置，发布，需要注意作业参数， $XXX_NAME使用$PROJDEF，就使用工程的参数，都一样，还是会自动替换，不是同一个变量，而是变量集
