#### Compliant 抱怨一下。。。想想解决方案呢？
说说感受把，搭建服务，提供服务这方是运维的相关的，确实很生疏，不太熟悉权限，为什么要这样子做，就像Docker一样，建立一个服务，每个服务都有它自己的特性，这样子使得搭建服务人员好累啊，都得看手册，属性才能完成任务。。

在浏览Google，[教程](http://blog.csdn.net/u011726984/article/details/52036654)的时候，其实提到Git是分布式的，所以也就没所谓的Server/Client端，只是用的时候才有一点区别，

在Windows上安装Git Server端，其实就需要两个程序，一个就是Git本身，他就包含了ServerClient端，因为他就是分布式的，另外一个是SSH服务，因为Git用到的一个传输方式就是SSH，因此需要SSH服务端，常用的就是Copssh，它安装的时候就需要建立用于ssh登录的用户名，用户目录，

注意要把git相关的程序exe和lib dll都要放到copssh目录下，因为ssh路径问题导致，

然后在ssh用户的目录下使用git init --bare 创建服务端的仓库，

然后在客户端上clone 或者建立自己仓库，然后 remote add origin - push ， 注意这里不成功原因可能是git服务器端要把仓库放在它的home目录，然后就应该没有问题了。

注意每个人在使用前应该:
```
git config --global user.name "YOUR NAME"
git config --global user.email "YOUR EMAIL ADDRESS"
```
